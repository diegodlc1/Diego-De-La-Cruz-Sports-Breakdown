## Dallas Mavericks Offseason Breakdown

The Mavericks made a massive trade at the deadline, sending Luke Doncic to the Lakers in exchange for AD. With this trade, it seems the Mavs will look to win now with this roster. ## Draft

The Mavs only have one draft pick in this year's draft. They own their own first-round pick coming in at the 11th pick. With this pick, this is the player and salary hit.

1. Thomas Sorber, 5.715 million hit for next season

If the defense wins championships, philosophy still stands, and Sorber would be the selection here. He's a great 2-way player who can develop behind AD. I see him being a PF more so than a Center, so I believe that Sorber would be a great selection. He's not ready to compete just yet, but being able to develop in that frontcourt, he could turn into a great player. It's not a win now move but it is a move
