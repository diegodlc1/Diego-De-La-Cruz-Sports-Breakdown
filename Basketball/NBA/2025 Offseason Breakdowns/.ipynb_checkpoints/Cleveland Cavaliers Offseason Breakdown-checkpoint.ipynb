## Extensions

This is where the bulk of the Cavs' work will be done. They're at 11 players making the salary on the table. They're well above the second apron, and I see no way the Cavs get under that unless they trade key pieces. Looking at possible extensions, Jerome and Merrill will be big ones, and Tristan Thompson is most likely coming back anyway on the vet minimum. Green and Tomlin aren't needed too much, so a full contract isn't necessary for them. Thompson will get the veteran minimum, and that's all we need to mention about him. The Cavaliers only have his Early-Bird rights and won't be able to sign him. The Cavs can only offer 4,481,706, he's going for close to the full MLE. Merril will be able to resign with the Cavaliers as they have his full Bird rights. With how strong the SG market is, I don't think Merrill will warrant big numbers from another team. He should be able to get around 9 million from the Cavaliers next season. I don't see a team offering more than that for him. 2-year 9 million a year isn't terrible. Tomlin can get a 2-way, cheap contract or a G-League offer from the Cavs. I'll give him a G-League offer for now.

## Trade and Free Agency.

Trading will be the only way the Cavs can improve in the margins. At the moment, they don't need to make rash decisions. They traded for Hunter at the deadline and a full offseason; adapting to those around him could help the Cavaliers elevate to the next level. Give them this offseason before they make big moves. To me, the Cavaliers only have 1 player they can use in a trade, Isaac Okoro. The only other player you can make an argument for is Dean Wade, but I'm not confident in Tomlin or Okeke's ability to play backup minutes in an already thin Cavaliers frontcourt.

As a Second Apron team, the Cavaliers can't combine players or take on more money than what they send out in a trade. The Cavs need a backup Center and a Backup PG to fill in depth there. Okoro is on a bad contract, but his age is still somewhat intriguing. I think a young tanking team can take a chance on him, or an old playoff team that needs to make changes can try to get him. I only see 3 teams with players the Cavs wouldn't mind. Chicago, Orlando, and Houston. It'll be Lonzo, Goga, or Holiday. If I'm the Cavs, I'm trying to get Goga. Is he young, and can he develop? Yes. Are the Magic desperate for shooters? Yes. Is Okoro the best you can get from trading Goga?
