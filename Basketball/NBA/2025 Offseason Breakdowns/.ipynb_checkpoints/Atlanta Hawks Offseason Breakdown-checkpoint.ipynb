# Atlanta Hawks Offseason Breakdown

The Atlanta Hawks are right there with the Kings. A team looking to compete but haven't gotten over the hump since the 2021 season. Truthfully, the Hawks have some pieces. While I don't believe that the Hawks can contend for a title, the playoffs shouldn't be out of the scope. I don't think that they need to tank; they have pieces now, but they're in the basketball purgatory spot at the moment. We can cough it up to injuries this year, but I don't think that with a healthy squad, they were on the same level as the playoff teams in the East outside of the Heat. With that being said, this is how I would go about the offseason.

## Draft

The Hawks have 2 first-rounders in the draft this year. They own the Kings' 13th pick and the Lakers' 22nd pick. They will be making both selections in this year's draft unless a team were to want to trade for the 22nd pick. Either way, I don't do draft day trades, so they make both selections in this case.

1. Jase Richardson, 4.9 million cap hit next season
2. Rasheer Fleming, 3.372 million cap hit next season

The Hawks need 2 things. Front-court depth and PG depth. I think they have solid wing players, but the frontcourt is severely lacking in players. Richardson is a solid prospect in his PG abilities. Able to create and score, they can develop him to work under Trae Young. Fleming is another interesting prospect in his ability to do everything well. He's a catch and shoot big who provides great spacing for whatever team gets him. This would be a solid draft for the Hawks and fits their system well. Add 2 players for 8.272 million.
