## Free Agents and Trades

Ok, this team is over the first apron and not a championship team. The Heat have 2 directions: blow it up and tank for the next 2 years, or make a championship push with what they have. Tanking means building around Herro and trading everyone over the age of 26. Get a high pick and open cap space, and make a push in 2 to 3 years. The other option is to retool around what you have. Honestly, both make some sense, but Miami is a free agent and trade destination. They can get stars. The thing is that both ways can allow that to happen. If I'm the Heat, I'm building off what I have. You can try and combine Robinson and Rozier for someone, but I think that their value will be higher at the deadline for a team looking for cap space in the offseason, if not, then the Heat will open around 46 million in cap space themselves. I don't think getting Markkanen is worth that unless the Heat add picks for a different player. Trade them at the deadline or ride out the year with them and open that cap space yourself. Wiggins will stay another year, and if you're retooling, Bam and Herro are staying. At this point, a trade will have to be made. The 2 highest candidates for that are Highsmith and Anderson. They need to get below the first apron, and these 2 are the most expendable with some value. Highsmith could net Eubanks from the Clippers, Powell from Dallas, Gordon from the Sixers, Champaignie from the Spurs, Milton from the Lakers, and Juzang from the Jazz. With that being said, the Clippers don't need Highsmith. Dallas won't trade with Miami. Gordon is the worst-case scenario for the Heat. Champaignie would be good value, but doesn't fit a need for the Heat, and I think the Spurs would want more than Highsmith and a second, something the Heat shouldn't give up on a salary dump. Lakers don't need Highsmith. I don't think the Jazz want to get off of Juzang with his deal. Anderson will pull a bigger market. Orlando, San Antonio, OKC, Memphis, and Denver have the contracts and need for the Heat to get Anderson. I think the Heat can get a disappointing young player on one of these teams. Looking at that, we can eliminate to 2 teams, OKC and Orlando. Orlando is pushing on the first apron and would need to drop 3.6 million to trade Jett Howard. This leaves just OKC, which leaves Ousmane Dieng for the trade. I would do it if I were Miami and OKC. Get off Anderson's contract, get under the Second Apron, and get an expiring young player to see what he has for the year, if OKC picks up his option, and decide what to do with him the following season.
